# **Add share=True**

```python
# This will patch the last line of app.py to add share=True
import fileinput

for line in fileinput.input("app.py", inplace=True):
    # Replace the launch line that currently does not have share=True
    if ".launch()" in line:
        # Add share=True if not already present
        if "share=" not in line:
            line = line.replace(".launch()", ".launch(share=True)")
    print(line, end="")
```
